```python
import os
import shutil

import numpy as np

# Define the source folders and their paths
source_folders = {
     "r431.romulus25cvdXsec.3072g1HsbBH": {'path': '/data/REPOSITORY/romulus_zooms/r431.romulus25cvdXsec.3072g1HsbBH'},
     "r492.romulus25cvdXsec.3072g1HsbBH": {'path': '/data/REPOSITORY/romulus_zooms/r492.romulus25cvdXsec.3072g1HsbBH'},
    # add files in /data/akaxia/rzooms
    "r634.romulus25si2s50v35": {'path': '/data/akaxia/rzooms/r634.romulus25si2s50v35.3072g1HsbBH'},
    "r492.romulus25si2s50v35": {'path': '/data/akaxia/rzooms/r492.romulus25si2s50v35.3072g1HsbBH'},
    "r468.romulus25si2s50v35": {'path': '/data/akaxia/rzooms/r468.romulus25si2s50v35.3072g1HsbBH'},
    "r488.romulus25si2s50v35": {'path': '/data/akaxia/rzooms/r488.romulus25si2s50v35.3072g1HsbBH'},
    "r544.romulus25si2s50v35": {'path': '/data/akaxia/rzooms/r544.romulus25si2s50v35.3072g1HsbBH'},
    "r597.romulus25si2s50v35": {'path': '/data/akaxia/rzooms/r597.romulus25si2s50v35.3072g1HsbBH'},
    "r523.romulus25si2s50v35": {'path': '/data/akaxia/rzooms/r523.romulus25si2s50v35.3072g1HsbBH'},
    "r618.romulus25si2s50v35": {'path': '/data/akaxia/rzooms/r618.romulus25si2s50v35.3072g1HsbBH'},
 }

#also add CDM equivalents in  source data/REPOSITORY/romulus_zooms
# 
source_folders = {
    #/data/akaxia/rzooms/r488.romulus25si2s50v35.3072g1HsbBH
    "r431.romulus25.3072g1HsbBH": {'path': '/data/REPOSITORY/romulus_zooms/r431.romulus25.3072g1HsbBH'},
    "r492.romulus25.3072g1HsbBH": {'path': '/data/REPOSITORY/romulus_zooms/r492.romulus25.3072g1HsbBH'},
    "r468.romulus25.3072g1HsbBH": {'path': '/data/REPOSITORY/romulus_zooms/r468.romulus25.3072g1HsbBH'},
    "r488.romulus25.3072g1HsbBH": {'path': '/data/REPOSITORY/romulus_zooms/r488.romulus25.3072g1HsbBH'},
    "r544.romulus25.3072g1HsbBH": {'path': '/data/REPOSITORY/romulus_zooms/r544.romulus25.3072g1HsbBH'},
    "r597.romulus25.3072g1HsbBH": {'path': '/data/REPOSITORY/romulus_zooms/r597.romulus25.3072g1HsbBH'},
    "r523.romulus25.3072g1HsbBH": {'path': '/data/REPOSITORY/romulus_zooms/r523.romulus25.3072g1HsbBH'},
    "r618.romulus25.3072g1HsbBH": {'path': '/data/REPOSITORY/romulus_zooms/r618.romulus25.3072g1HsbBH'},
    "r634.romulus25.3072g1HsbBH": {'path': '/data/REPOSITORY/romulus_zooms/r634.romulus25.3072g1HsbBH'},
}



# Define the destination base directory
dest_base_dir = '/home/bk639/data/MerianSIDM'

import os
from collections import defaultdict

import os
from collections import defaultdict

import os
from collections import defaultdict

def create_symlinks(source_path, dest_path, exclude_dirs=None):
    if exclude_dirs is None:
        #also filter noL noboost_loweff photogenic_finding ahf_100 noL_loweff_noboost noL_higheff_noboost L_higheff_boost noBHs L_higheff_noboost
        exclude_dirs = ["ahf_200", "shield","adiabatic","orig", "noL", "noboost_loweff", "photogenic_finding", "ahf_100", "noL_loweff_noboost", "noL_higheff_noboost", "L_higheff_boost", "noBHs", "L_higheff_noboost"]
    
    # Dictionary to store file names and their locations
    file_locations = defaultdict(list)
    
    # Walk through all subdirectories
    for root, dirs, files in os.walk(source_path):
        # Remove excluded directories from the dirs list
        dirs[:] = [d for d in dirs if d not in exclude_dirs]
        
        for file in files:
            # Get the full path of the source file
            source_file = os.path.join(root, file)
            
            # Store the file location
            file_locations[file].append(source_file)

    # Dictionary to store duplicate files
    duplicates = {file: locations for file, locations in file_locations.items() if len(locations) > 1}

    # Create symlinks
    for file, locations in file_locations.items():
        source_file = locations[0]  # Use the first occurrence of the file
        dest_file = os.path.join(dest_path, file)

        # If a file with the same name already exists in the destination, add a suffix
        counter = 1
        while os.path.exists(dest_file):
            base, ext = os.path.splitext(file)
            dest_file = os.path.join(dest_path, f"{base}_{counter}{ext}")
            counter += 1

        # Create the symlink
        os.symlink(source_file, dest_file)
        #print(f"Created symlink: {dest_file} -> {source_file}")

    # Report duplicate files
    if duplicates:
        print("\nDuplicate files found:")
        for file, locations in duplicates.items():
            print(f"\n{file}:")
            for location in locations:
                print(f"  - {location}")

    return duplicates

# Main execution
for folder_name, folder_info in source_folders.items():
    source_path = folder_info['path']
    dest_path = os.path.join(dest_base_dir, folder_name)

    # Create the destination folder if it doesn't exist
    os.makedirs(dest_path, exist_ok=True)

    # Create symlinks
    create_symlinks(source_path, dest_path)

print("Symlink creation completed.")
```

```python
import os
import pandas as pd
from pathlib import Path

# Set the directory to check
directory = '/home/bk639/data/MerianSIDM'

def is_symlink_folder(path):
    return os.path.islink(path) and os.path.isdir(path)

def is_symlink_file(path):
    return os.path.islink(path) and os.path.isfile(path)

def find_symlinks(directory):
    symlinked_folders = []
    symlinked_files = []
    
    for root, dirs, files in os.walk(directory):
        for dir_name in dirs:
            full_path = os.path.join(root, dir_name)
            if is_symlink_folder(full_path):
                symlinked_folders.append({
                    'path': full_path,
                    'target': os.readlink(full_path)
                })
        
        for file_name in files:
            full_path = os.path.join(root, file_name)
            if is_symlink_file(full_path):
                symlinked_files.append({
                    'path': full_path,
                    'target': os.readlink(full_path)
                })
    
    return symlinked_folders, symlinked_files

# Find symlinked folders and files
symlink_folders, symlink_files = find_symlinks(directory)

# Create DataFrames from the results
df_folders = pd.DataFrame(symlink_folders)
df_files = pd.DataFrame(symlink_files)

# Display the results for folders
print("Symlinked Folders:")
if df_folders.empty:
    print("No symlinked folders found.")
else:
    print(f"Found {len(df_folders)} symlinked folder(s):")
    display(df_folders)

print("\nSymlinked Files:")
if df_files.empty:
    print("No symlinked files found.")
else:
    print(f"Found {len(df_files)} symlinked file(s):")
    display(df_files)

# Optional: Save the results to CSV files
# df_folders.to_csv('symlinked_folders.csv', index=False)
# df_files.to_csv('symlinked_files.csv', index=False)
```
